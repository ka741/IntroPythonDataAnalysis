# Introduction

**What is exploratory data analysis?**

Exploratory Data Analysis describes a handful of statistical and visualization methods we can use to understand our data. Through these methods we can discover the major characteristics of our dataset. This information is important for determining  what statistical analysis is the best fit for our research question.

**How are we exploring data today?**

We will be using a dataset collected during the global pandemic. This data includes measures of social distancing and antisocial behavior. The original publication "Reduced social distancing early in the COVID-19 pandemic is associated with antisocial behaviors in an online United States sample" can be found here https://pubmed.ncbi.nlm.nih.gov/33412567/ and the csv file we will be using can be found here https://osf.io/4yp23/.

**What question are we trying to answer?**

Did people that are generally more antisocial (i.e. people that break the rules and are more physically and socially aggressive) leave their house more, violating government safety guidelines, during the global pandemic?
We can use publicly available data from the dataset above to answer this and plenty of other questions!

## Correlation Table

**What is a correlation table?**

Correlation tables give a numeric value to the strength and direction of a relationship between two variables.

**Strength**: Correlations range from 01. If two variables are not related the correlation will be closer to 0 (e.g. heart rate and income). The closer to 1 the correlation is the stronger the relationship (e.g. height and weight).

**Direction**: Correlations can also be positive or negative. If as one variable increases so does the other you will get a positive correlation (e.g. height and weight). If as one variable increases the other decreases you will get a negative correlation (e.g. supply and demand).

```
c = analyze.corr()
display(c)
```

Let's break down the line of code

> *c =* <br> 
*This is the name we are giving our correlation table. We are saying that from now on **c** is going to be shorthand for our correlation table.* 
<br> <br> 
*analyze* <br> 
*This is the DataFrame with the columns we want to calculate correlation values for.*
<br> <br> 
*.corr()* <br> 
*This calls the **corr** method to calculate correlation values between all the variables in our DataFrame.*

**How can we use correlations?**

This is a good place to have a "logic check". By this I mean, we know from previous research that some variables are associated in a certain direction.

We can see that people that are more physically aggressive are more socially aggressive (.92) and break rules more (.94). This makes sense!

If we look at our data and don't find that relationship it may be due to one of a few possibilities.


*   Data collection error
*   Coding or scoring error
*   We have captured a unique population

Any of these options is important to identify. It may mean you need to collect new data, recode your data, or pay special attention to what is different about the population you collected.

## Visualize Data with a ScatterPlot

***What is a scatter plot?***

Scatter plots display the relationship between two variables. It allows us to visualize concepts similar to the ones captured in the numeric correlation value (i.e. strength and direction). Stronger relationhsips will be more tightly packed and look more like a line. We can also view the direction, does one variable increase as the other decreases or do they go in the same direction?

```
sns.scatterplot(data=analyze, x="times_left_house_pastweek", y="STAB_social")
```

Let's break down the line of code

> *sns.scatterplot* <br> 
*This calls the seaborn library using sns then specifically calls the **scatterplot** method turns our variables into a scatterplot.* 
<br> <br> 
*(data=analyze, x="times_left_house_pastweek", y="STAB_social")* <br> 
*These are your parameters, **data** is specifying the DataFrame you are pulling the columns from, **x** is specifying the variable you are plotting on the x-axis, **y** is specifying the variable you are plotting on the y-axis. Make sure to put your column names in quotation marks.*

**How can we use a scatterplot?**

A scatter plot lets us visualize potential relationships between variables. It can help us visualize what the correlation values mean.
That being said, you don't need a correlation value to interpret a scatter plot.

In this data the direction suggests that more socially aggressive people tend to leave the house more. However, the relationhsip doesn't look very strong, meaning the dots are spread out and don't make a clear line.
